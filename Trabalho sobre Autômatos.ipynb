### Classe que montado o autômato

```python
import re
class Automato:
    def __init__(self, alfabeto):
        self.alfabeto = alfabeto
        self.estado_inicial = ''
        self.estados_finais = []
        self.transicoes = {}
    
    def definirEstadoInicial(self, nomeEstado):
        self.estado_inicial = nomeEstado

    def definirEstadoFinal(self, nomeEstado):
        self.estados_finais.append(nomeEstado)
    
    def inserirTransicao(self, estadoOrigem, simbolo, estadoDestino):
        try:
            re.match(self.alfabeto, simbolo)
            
            if estadoOrigem in self.transicoes:
                self.transicoes[estadoOrigem][simbolo] = estadoDestino
            else:
                self.transicoes.update({estadoOrigem: {simbolo: estadoDestino}})
            return True

        except:
            if simbolo == None and estadoDestino == None:
                    self.transicoes.update({estadoOrigem: {}})
                    return True
            else:
                return False
```

### Classe que valida uma cadeia de acordo com o autômato passado

```python
class Validacao:
    def __init__(self, transicoes, estado_inicial ,estados_finais, cadeias):
        self.transicoes = transicoes
        self.estado_inicial = estado_inicial
        self.estados_finais = estados_finais
        self.cadeias = cadeias
        self.resultado = ''
        
        self.__str__()
    
    def validarCadeia(self, cadeia):
        estado = self.estado_inicial
        try:
            for x in cadeia:
                estado = self.transicoes[estado][x]
            if (estado in self.estados_finais):
                self.resultado = 'Aceito'
            else:
                self.resultado = 'Rejeitado'
        except:
            self.resultado = 'Rejeitado'
    
    def __str__(self):
        for i in range(len(self.cadeias)):
            self.validarCadeia(self.cadeias[i])
            print(f'Cadeia: {self.cadeias[i]:<10}, {self.resultado}')
```

## **3 - Implemente um autômato finito que reconheça todas as ocorrências da palavra computador no texto T. O programa deve apontar em quais posições ocorreram o casamento exato da palavra.**

### Classe que valida uma palavra de acordo com autômato passado

```python
class BuscarPalavra:
    def __init__(self, transicoes, estado_inicial ,estados_finais, texto):
        self.transicoes = transicoes
        self.estado_inicial = estado_inicial
        self.estados_finais = estados_finais
        self.texto = texto
        self.posicoes = []
        
        self.__str__()
    
    def buscar(self):
        for i in range(len(self.texto)):
            estado = self.estado_inicial  
            try:
                for x in self.texto[i][1]:
                    estado = self.transicoes[estado][x]
                if estado in self.estados_finais:
                    self.posicoes.append(self.texto[i][0])
            except:
                pass
    
    def __str__(self):
        self.buscar()
        if len(self.posicoes) == 0:
            print('Não foi encontrada nenhuma ocorrência da palavra')
        else:
            for i in range(len(self.posicoes)):
                print(f'Palavra encontrada no índice: {self.posicoes[i]}')
```

### Método para dividir um texto em palavras limitadas por um espaço

```python
import re

def dividirTexto(texto):
    palavras = []
    for m in re.finditer(r'\S+', texto):
        palavras.append([m.start(), m.group()])
    return palavras
```

## **4 - Implemente um transdutor finito (máquina de Moore ou Mealy) que, dada uma sequência de moedas de 25 e 50 centavos e de 1 real, forneça uma lata de refrigerante quando a sequência totalizar 1 real ou mais. Cada moeda inserida deverá corresponder a uma de duas saídas: 0, se uma lata não pode ser (ainda) liberada, ou 1, se uma lata deve ser liberada.**

```python
class MaquinaRefrigerante:
    def __init__(self, transicoes, estado_inicial ,estados_finais, moedas):
        self.transicoes = transicoes
        self.estado_inicial = estado_inicial
        self.estados_finais = estados_finais
        self.moedas = moedas
        self.resultado = ''
        self.estado = 'q0'
        self.__str__()
    
    def inserirMoeda(self, moeda):
        try:
            movimento = self.transicoes[self.estado][moeda]
            self.estado = movimento[0]

            if (self.estado in self.estados_finais) and (movimento[1] == 1):
                self.resultado = 1
            else:
                self.resultado = 0
        except:
            self.resultado = 0
    
    def __str__(self):
        for i in range(len(self.moedas)):
            self.inserirMoeda(self.moedas[i])
            if self.resultado == 1:
                print(f'Moeda inserida: {self.moedas[i]:<5}, {self.resultado} Retirar refrigerante')
            else:
                print(f'Moeda inserida: {self.moedas[i]:<5}, {self.resultado}')
```
