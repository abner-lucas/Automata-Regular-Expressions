## **1 - Implemente AFDs, em uma linguagem de programação à sua escolha, que aceitem as seguintes cadeias:**

```python
class Valida_AFD:
    def __init__(self, transicoes, estado_inicial ,estados_finais, cadeias):
        self.transicoes = transicoes
        self.estado_inicial = estado_inicial
        self.estados_finais = estados_finais
        self.cadeias = cadeias
        self.resultado = ''
        
        self.__str__()
    
    def validarCadeia(self, cadeia):
        estado = self.estado_inicial
        try:
            for x in cadeia:
                estado = self.transicoes[estado][x]
            if (estado in self.estados_finais):
                self.resultado = 'Aceito'
            else:
                self.resultado = 'Rejeitado'
        except:
            self.resultado = 'Rejeitado'
    
    def __str__(self):
        for i in range(len(self.cadeias)):
            self.validarCadeia(cadeias[i])
            print(f'Cadeia: {cadeias[i]:<10}, {self.resultado}')
```

```python
AFD = {0:{'0':0, '1': 1},
       1:{'0':2},
       2:{'0':0}}
estado_inicial = 0
estados_finais = {0}
cadeias = ['0','0100','100','1','1001','010','0001']

resultado = Valida_AFD(AFD, estado_inicial, estados_finais, cadeias)
```

## **3 - Implemente um autômato finito que reconheça todas as ocorrências da palavra computador no texto T. O programa deve apontar em quais posições ocorreram o casamento exato da palavra.**

## T = “O **computador** é uma máquina capaz de variados tipos de tratamento automático de informações ou processamento de dados. Entende-se por **computador** um sistema físico que realiza algum tipo de computação. Assumiu-se que os computadores pessoais e laptops são ícones da era da informação. O primeiro **computador** eletromecânico foi construído por Konrad Zuse (1910–1995). Atualmente, um microcomputador é também chamado **computador** pessoal ou ainda computador doméstico.”

```python
import re

class BuscarPalavra:
    def __init__(self, transicoes, estado_inicial ,estados_finais, texto):
        self.transicoes = transicoes
        self.estado_inicial = estado_inicial
        self.estados_finais = estados_finais
        self.texto = texto
        self.posicoes = []
        
        self.__str__()
    
    def buscar(self):
        for i in range(len(self.texto)):
            estado = self.estado_inicial  
            try:
                for x in self.texto[i][1]:
                    estado = self.transicoes[estado][x]
                if estado in self.estados_finais:
                    self.posicoes.append(self.texto[i][0])
            except:
                pass
    
    def __str__(self):
        self.buscar()
        if len(self.posicoes) == 0:
            print('Não foi encontrada nenhuma ocorrência da palavra')
        else:
            for i in range(len(self.posicoes)):
                print(f'Palavra encontrada no índice: {self.posicoes[i]}')

def dividirTexto(texto):
    palavras = []
    for m in re.finditer(r'\S+', texto):
        palavras.append([m.start(), m.group()])
    return palavras
```

```python
AFD = {0:{'c':1},
       1:{'o':2},
       2:{'m':3},
       3:{'p':4},
       4:{'u':5},
       5:{'t':6},
       6:{'a':7},
       7:{'d':8},
       8:{'o':9},
       9:{'r':10},
       10:{}}

estado_inicial = 0
estados_finais = {10}
texto = 'O computador é uma máquina capaz de variados tipos de tratamento automático de informações ou processamento de dados. Entende-se por computador um sistema físico que realiza algum tipo de computação. Assumiu-se que os computadores pessoais e laptops são ícones da era da informação. O primeiro computador eletromecânico foi construído por Konrad Zuse (1910–1995). Atualmente, um microcomputador é também chamado computador pessoal ou ainda computador doméstico.'

palavras = dividirTexto(texto)

resultado = BuscarPalavra(AFD, estado_inicial, estados_finais, palavras)
```

## **4 - Implemente um transdutor finito (máquina de Moore ou Mealy) que, dada uma sequência de moedas de 25 e 50 centavos e de 1 real, forneça uma lata de refrigerante quando a sequência totalizar 1 real ou mais. Cada moeda inserida deverá corresponder a uma de duas saídas: 0, se uma lata não pode ser (ainda) liberada, ou 1, se uma lata deve ser liberada.**

```python
class MaquinaRefrigerante:
    def __init__(self, transicoes, estado_inicial ,estados_finais, moedas):
        self.transicoes = transicoes
        self.estado_inicial = estado_inicial
        self.estados_finais = estados_finais
        self.moedas = moedas
        self.resultado = ''
        self.estado = 0
        self.__str__()
    
    def inserirMoeda(self, moeda):
        try:
            movimento = self.transicoes[self.estado][moeda]
            self.estado = movimento[0]

            if (self.estado in self.estados_finais) and (movimento[1] == 1):
                self.resultado = 1
            else:
                self.resultado = 0
        except:
            self.resultado = 0
    
    def __str__(self):
        for i in range(len(self.moedas)):
            self.inserirMoeda(self.moedas[i])
            if self.resultado == 1:
                print(f'Moeda inserida: {self.moedas[i]:<5}, {self.resultado} Retirar refrigerante')
            else:
                print(f'Moeda inserida: {self.moedas[i]:<5}, {self.resultado}')
```
