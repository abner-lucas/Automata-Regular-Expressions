## **Validação de campos**

```python
import re

# em regex [a-z] = Conjunto Sigma
# em regex [A-Z] = Conjunto Gamma
# em regex \d = [0-9] = Conjunto N

campos = ['nome', 'e-mail', 'senha', 'CPF', 'telefone', 'data e horário', 'número real']
```

```python
class Validacao:
    def __init__(self, expressao, cadeias, campo):
        self.cadeias = cadeias
        self.expressao = expressao
        self.campo = campo
        self.mensagem = ''
        
        self.__str__()

    def validarCadeia(self, cadeia, expressao):
        resultado = re.match(expressao, cadeia)
        if resultado != None:
            self.mensagem = 'Aceito'
        else:
            self.mensagem = 'Rejeitado'
        return self.mensagem

    def __str__(self):
        for i in range(len(self.cadeias)):
            self.validarCadeia(self.cadeias[i], self.expressao)
            print(f'{i+1:>2} - {self.campo}: {self.cadeias[i]:20}, {self.mensagem}')
```

## Nome
Deve receber o **nome** e o **sobrenome**, ambos não vazios, separados por um espaço;
não deve aceitar caracteres especiais ou numéricos; o primeiro símbolo do nome e do
sobrenome deve ser do alfabeto Γ, e os outros símbolos devem ser do alfabeto Σ.
- Ex. de cadeias aceitas: Alan Turing, Noam Chomsky, Ada Lovelace
- Ex. de cadeias rejeitadas: 1Alan, Alan, A1an, A1an Turing

```python
nomes = ['Alan Turing', 'Noam Chomsky', 'Ada Lovelace', '1Alan', 'Alan', 
        '', 'N Chomsky', 'Alan T', 'A1an', 'A1an Turing']

regexNome = r'(^[A-Z][a-z]*\s[A-Z][a-z]*$)+'

resultado = Validacao(regexNome, nomes, campos[0])
```

## E-mail
As sentenças possuem símbolos de Σ e deve conter exatamente um símbolo “@”; não devem começar com o símbolo “@”; devem terminar com a sequência “.br”; devem ter, pelo menos, um símbolo de Σ entre o símbolo “@” e o “.br”
- Ex. de cadeias aceitas: a@a.br, divulga@ufpa.br
- Ex. de cadeias rejeitadas: @, a@.br, T@teste.br

```python
emails = ['', 'site.domain.br', 'a@a.br', 'divulga@ufpa.br',
        '@', 'a@.br', 'ggglll@domain.br', 'T@teste.br',
        'address@domain.br', '@test@domain.br']

regexEmail = r'(^[a-z]+@[a-z]+(.br)$)+'

resultado = Validacao(regexEmail, emails, campos[1])
```

## Senha
As cadeias podem conter símbolos dos alfabetos Σ, Γ e N; devem, obrigatoriamente, ter pelo menos um símbolo do alfabeto Γ e um símbolo do alfabeto N; devem ter comprimento igual a 8;
- Ex. de cadeias aceitas: 518R2r5e, F123456A, 1234567T, ropsSoq0
- Ex. de cadeias rejeitadas: F1234567A, abcdefgH, 1234567HI

```python
senhas = ['518R2r5e', 'F123456A', '1234567T', 'ropsSoq0',
          'F1234567A', 'abcdefgH', '1234567HI', '',
          '123456', 'akjg9wu3']

#regexSenha = r'^(?=(.*?[A-Z])+)(?=(.*?[\d])+)(?=(.*?[a-z])*).{8}$'
regexSenha = r'^((?=([A-Z]+([a-z]|[0-9]))|(([a-z]|[0-9])+[A-Z]))(?=([A-Z]|[a-z]|[0-9])+[0-9])([A-Z]|[a-z]|[0-9]){8})$'

resultado = Validacao(regexSenha, senhas, campos[2])
```

## CPF
As cadeias devem ter o formato xxx.xxx.xxx-xx, onde x ∈ N.
- Ex. de cadeias aceitas: 123.456.789-09, 000.000.000-00
- Ex. de cadeias rejeitadas: 123.456.789-0, 111.111.11-11

```python
cpfs = ['123.456.789-09', '00000000000', '123.456.789-0', '111.111.11-11',
       'ab0.000.000-00', '078.412.11423','000.000.000-00', '',
       '111111.11-11', '11.1.111.11-11']

regexCpf = r'(^([0-9]{3}[.]){2}[0-9]{3}[-][0-9]{2}$)+'

resultado = Validacao(regexCpf, cpfs, campos[3])
```

## Telefone
As cadeias devem ter o formato<br/>
(xx) 9xxxx-xxxx ou<br/>
(xx) 9xxxxxxxx ou<br/>
xx 9xxxxxxxx<br/>
onde x ∈ N.
- Ex. de cadeias aceitas: (91) 99999-9999, (91) 999999999, 91 999999999
- Ex. de cadeias rejeitadas: (91) 59999-9999, 99 99999-9999, (94)95555-5555

```python
telefones = ['(91) 99999-9999', '(91) 999999999', '91 999999999', '(91) 59999-9999',
       '99 99999-9999', '9999999-9999','(94)95555-5555', '',
       '95555-5555', '91955555555']

regexTel = r'(^([(][0-9]{2}[)]\s[9][0-9]{4}[-][0-9]{4})|([(][0-9]{2}[)]\s[9][0-9]{4}[0-9]{4})|([0-9]{2}\s[9][0-9]{4}[0-9]{4})$)+'

resultado = Validacao(regexTel, telefones, campos[4])
```

## Data e horário
As cadeias devem ter o formato dd/mm/aaaa hh:mm:ss, onde d, m, a, h, m, s ∈ N.
- Ex. de cadeias aceitas: 31/08/2019 20:14:55, 99/99/9999 23:59:59
- Ex. de cadeias rejeitadas: 99/99/9999 3:9:9, 9/9/99 99:99:99

```python
dts_hrs = ['31/08/2019 20:14:55', '99/99/9999 23:59:59', '99/99/9999 3:9:', '9/9/99 99:99:99',
           '', '99-08-2019 99:99:99', '9999/99/99 99:99:99', '99:99:99 31/08/2019',
           '1/1/99 20:14:55', '99/99/9999 20:14']

regexDtHr = r'(^[0-9]{2}[/][0-9]{2}[/][0-9]{4}\s[0-9]{2}[:][0-9]{2}[:][0-9]{2}$)+'

resultado = Validacao(regexDtHr, dts_hrs, campos[5])
```

## Número real com ou sem sinal
As cadeias devem começar com um dos símbolos do alfabeto {+, -, ε}; depois, pelo menos, um símbolo do alfabeto N; depois, exatamente, um símbolo separador {“.”}; para finalizar, pelo menos um símbolo do alfabeto N; exceção: números sem a parte fracionária também devem ser aceitos.
- Ex. de cadeias aceitas: -25.467, 1, -1, +1, 64.2
- Ex. de cadeias rejeitadas: 1., .2, +64,2

```python
num = ['-25.467', '1', '-1', '+1', '64.2', '1.', '.2', '+64,2', '+25.467', '-.4']

regexNum = r'(^(([+-]{0,1}[0-9]{1,}[.][0-9]+)|([+-]{0,1}[0-9]{1,}))$)+'

resultado = Validacao(regexNum, num, campos[6])
```
